```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generate random numbers from a skewed normal distribution
x=np.random.standard_t(10, size=8000)
x[:15]
```

# Range

```python
maxi=np.max(x)
mini=np.min(x)
```

```python
print(f"maximum value is: {maxi}")
print(f"minimum value is: {mini}")
```

```python
rangee=maxi-mini
print(f"range is: {rangee}")
```

# Inter Quartile Range (IQR)

```python
q1=np.percentile(x,25)
q3=np.percentile(x,75)
q1
```

```python
iqr=q1-q3
print("IQR: ",iqr)
```

# Mean Absolute Deviation(MAD)

```python
mean=x.mean()
```

```python
mad=np.mean(np.abs(x-mean))
print("MAD: ",mad)
```

```python
plt.hist(x, color='blue')
plt.axvline(x=mad,color='red')
plt.axvline(x=mean,color='yellow')
```

# Uniform Distribution

```python
# Set the seed for reproducibility
np.random.seed(42)

# Generate samples from a uniform distribution (0,1)
population = np.random.uniform(0,1,size=100000) #range(0,1)

#Plot the population distribution
plt.figure(figsize=(12,6))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.6)
plt.title("Population Distribution (Uniform)")
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
type(population)
```

```python
np.mean(population)
```

```python
# Sample sizes to consider
samplesize=[5,10,30,100,1000]

# Generate samples of different sizes and calculate sample means
samplemeans=[]
for n in samplesize:
    samples=np.random.choice(population,size=(1000,n),replace=True)
    means=np.mean(samples,axis=1)
    print(type(means))
    samplemeans.append(means)
# Plot the distribution of sample means for each sample size
plt.figure(figsize=(12,8))
for i,means in enumerate (samplemeans):
    plt.subplot(2,3,i+1)
    plt.hist(means,bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"sample size: {samplesize[i]}")
    plt.xlabel('sample mean')
    plt.ylabel('density')
plt.tight_layout()
plt.show()
```
