```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generate random numbers from a skewed normal distribution
x=np.random.standard_t(10, size=8000)
x[:15]
```

```python
plt.hist(x,color='blue')
```

# Mean

```python
mean=np.mean(x)
print("Mean: ",mean)
```

```python
plt.hist(x,color='blue')
plt.axvline(x=mean,color='yellow')
```

# Median

```python
median=np.median(x)
print("Median: ",median)
```

```python
plt.hist(x,color='blue')
plt.axvline(x=median,color='yellow')
```

# Mode

```python
mode =stats.mode(x).mode
```

```python
plt.hist(x,color='blue')
plt.axvline(x=mode,color='yellow')
```

# Variance
Deflection

```python
variance=np.var(x)
print("Variance: ",variance)
```

```python
plt.hist(x,color='blue')
plt.axvline(x=mean,color='yellow')
plt.axvline(x=variance,color='red')
```

# Standard Deviation

```python
std=np.std(x)
print("Standard Deviation: ",std)
```

```python
plt.hist(x, color='blue')
plt.axvline(x=mean,color='yellow')
plt.axvline(x=variance,color='red')
plt.axvline(x=std,color='green')
```

# Mean Absolute Deviation(MAD)

```python
mean=np.mean(x)
mad=np.mean(np.abs(x-mean))
print("MAD: ",mad)
plt.hist(x, color='blue')
plt.axvline(x=mad,color='yellow')
plt.axvline(x=variance,color='red')
plt.axvline(x=std,color='green')
plt.axvline(x=mean,color='white')
```
