```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exponential Distribution

```python
# Set the seed for reproducibility
np.random.seed(42)

# Generate samples from an exponential distribution
population=np.random.exponential(scale=1, size=100000)

# Plot the population distibution
plt.figure(figsize=(8,6))
plt.hist(population,bins=50,density=True, color='blue', alpha=0.6)
plt.title("Population Distribution (Exponential)")
plt.xlabel('value')
plt.ylabel('Density')
plt.xlim(0,5)
plt.show()
```

```python
# Sample sizes to consider
samplesize=[5,10,30,100,1000,7]

# Generate samples of different sizes and calculate sample means
samplemeans=[]
for n in samplesize:
    samples=np.random.choice(population,size=(1000,n),replace=True)
    means=np.mean(samples,axis=1)
    print(type(means))
    samplemeans.append(means)
# Plot the distribution of sample means for each sample size
plt.figure(figsize=(12,8))
for i,means in enumerate (samplemeans):
    plt.subplot(2,3,i+1)
    plt.hist(means,bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"sample size: {samplesize[i]}")
    plt.xlabel('sample mean')
    plt.ylabel('density')
plt.tight_layout()
plt.show()
```

# Hypothesis Testing
what is pdf and cdf

```python
from scipy import stats
```

```python
# Generate x values for the standard normal distribution
x=np.linspace(-4, 4, 1000)
plt.plot(x)

# Calculate the critical z-scores
# ppf = percent point function
z_critical=stats.norm.ppf(1 - 0.05/2)

# Plot the standard normal distribution
plt.figure(figsize=(8,6))
plt.plot(x, stats.norm.pdf(x), 'black', lw=2, label='Standard Normal Distribution')

# Shade the rejection region
plt.fill_between(x, stats.norm.pdf(x), where=(x <= -z_critical) | (x >= z_critical),
                 color='red', alpha=0.3, label='Rejection Region')

# Add labels and title
plt.title('Standard Normal Distribution with Rejection Region')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.legend()
plt.show()
```
